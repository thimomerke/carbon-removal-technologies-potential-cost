Build a plot for the cost of carbon removal for different technologies

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
cost_data = pd.read_excel("cost.xlsx")
cost_data.head()
```

```python
spec = dict(x="tech", y="price", data=cost_data)

plt.figure(figsize=(20,6))
sns.set(font_scale=1.2)
sns.color_palette("tab10")

plot = sns.stripplot(**spec, size=4, jitter=False, s=10, linewidth=1, alpha=.5, hue="Source")
interval = sns.pointplot(**spec, join=False, ci=0, capsize=.1, scale=0)
hundred = plt.axhline(y=100, color='r', linestyle='-')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=5)
plot.set(xlabel ="Technology", ylabel = "Cost per t/CO2 removed in USD")
```
